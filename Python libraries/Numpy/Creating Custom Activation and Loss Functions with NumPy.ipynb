Creating Custom Activation and Loss Functions with NumPy

```python
import numpy as np
```

```python
#   ------------------------------------ Sigmoid Function ------------------------------------
# The sigmoid function takes any number between 0 and 1, which helps in finding probabilities or chances.

# Random data
x = np.array([-2.0, 0.0, 2.0])

# Create the sigmoid function (using np.exp for the exponential calculation)
sigmoid = 1 / (1 + np.exp(-x))

print("Sigmoid Output:", sigmoid)
```

```python
#  ------------------------------------ Log Loss (Binary Cross-Entropy) ------------------------------------

# Log Loss, also known as Binary Cross-Entropy, is a loss function used for binary classification problems. 
# It measures the performance of a classification model whose output is a probability value between 0 and 1. The formula for log loss is:
# Log Loss = - (y * log(p) + (1 - y) * log(1 - p))

# let, actual label is 1 and predicted probability is 0.9, then the log loss can be calculated as follows:
true_label = 1
prediction = 0.9

# Calculate log loss
loss = - (true_label * np.log(prediction) + (1 - true_label) * np.log(1 - prediction))

print("Log Loss:", loss)
# Output: 0.105... (the lower, the better the model's performance)
```

```python
#  ------------------------------------ ReLU Function ------------------------------------
# ReLU (Rectified Linear Unit) is an activation function that outputs the input directly if it is positive; 
# otherwise, it outputs zero. The formula for ReLU is:


inputs = np.array([-3.5, -1.2, 0.0, 2.8, 5.1])

# ReLU function creation (using np.maximum)
relu_output = np.maximum(0, inputs)

print("ReLU Output:", relu_output) # negative values are set to 0, positive values remain unchanged
```

```python
#  ------------------------------------ Tanh Function ------------------------------------
# This is an activation function similar to the sigmoid,
# but it maps the data between -1 and 1 instead of 0 and 1. The formula for tanh is:
# Tanh(x) = (e^x - e^(-x)) / (e^x + e^(-x))

x_vals = np.array([-2.0, 0.0, 2.0])

# Tanh function creation (using np.tanh for the hyperbolic tangent calculation)
tanh_output = np.tanh(x_vals)

print("Tanh Output:", tanh_output)
```
