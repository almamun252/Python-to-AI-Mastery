Basic Calculation

```python
import numpy as np

# Creating a NumPy array from a list of scores
scores = np.array([85, 92, 78, 99, 88])
print("Scores Array:", scores)

# Calculating the total sum of the scores
total = np.sum(scores)
print("Total Sum:", total) 

# Calculating the average score
average = np.mean(scores)
print("Average Score:", average) 
```

```python
#          ----------------------------------          Standard deviation           -------------------------------
# In Python, standard deviation is a measure of how spread out the values in a dataset are from the mean. It is calculated using the formula:
# std = sqrt(sum((x_i - mean)^2) / n)
# where x_i represents each value in the dataset, mean is the average of the dataset, and n is the number of values in the dataset.

# Calculating the standard deviation of the scores
std_dev = np.std(scores)
print("Standard Deviation:", std_dev) 
```

```python
#          ----------------------------------          Minimum and Maximum Values           -------------------------------

lowest = np.min(scores) # Finding the minimum score
highest = np.max(scores) # Finding the maximum score

print("Lowest Score:", lowest)   
print("Highest Score:", highest) 
```

```python
#       ----------------------------------          Index of Minimum and Maximum Values           -------------------------------

# Finding the index of the minimum and maximum scores
min_index = np.argmin(scores)
max_index = np.argmax(scores)

print("Index of Lowest Score:", min_index)  
print("Index of Highest Score:", max_index) 
```

```python
# Working with 2D arrays (matrices) we can specify the axis along which to perform the sum.
matrix = np.array([[10, 20], 
                   [30, 40]])

print("Sum along columns (axis=0):", np.sum(matrix, axis=0)) 
print("Sum along rows (axis=1):", np.sum(matrix, axis=1))  

# Similarly, we can calculate the mean along different axes.
print("Mean along columns (axis=0):", np.mean(matrix, axis=0)) 
print("Mean along rows (axis=1):", np.mean(matrix, axis=1))

# We can also calculate the standard deviation along different axes.
print("Standard Deviation along columns (axis=0):", np.std(matrix, axis=0)) 
print("Standard Deviation along rows (axis=1):", np.std(matrix, axis=1))

# For 2D arrays, we can also find the minimum and maximum values along different axes.
print("Minimum along columns (axis=0):", np.min(matrix, axis=0)) 
print("Maximum along columns (axis=0):", np.max(matrix, axis=0))

print("Minimum along rows (axis=1):", np.min(matrix, axis=1))
print("Maximum along rows (axis=1):", np.max(matrix, axis=1))

# We can also find the indices of the minimum and maximum values along different axes.
print("Index of Minimum along columns (axis=0):", np.argmin(matrix, axis=0))
print("Index of Maximum along columns (axis=0):", np.argmax(matrix, axis=0))

print("Index of Minimum along rows (axis=1):", np.argmin(matrix, axis=1))
print("Index of Maximum along rows (axis=1):", np.argmax(matrix, axis=1))
```

```python
# For 3D arrays, we can specify the axis along which to perform the operations in a similar way.

arr3d = np.array([[[1, 2], [3, 4]],
                   [[5, 6], [7, 8]]])

# Sum along different axes for 3D array
print("Sum along axis=0:", np.sum(arr3d, axis=0))
print("Sum along axis=1:", np.sum(arr3d, axis=1))
print("Sum along axis=2:", np.sum(arr3d, axis=2))

# Mean along different axes for 3D array
print("Mean along axis=0:", np.mean(arr3d, axis=0))
print("Mean along axis=1:", np.mean(arr3d, axis=1))
print("Mean along axis=2:", np.mean(arr3d, axis=2))

# These ways of specifying the axis allow us to perform calculations along specific dimensions of our data,
# which is especially useful when working with multi-dimensional arrays.
```
