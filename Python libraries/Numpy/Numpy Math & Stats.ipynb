Basic Calculation

```python
import numpy as np

# Creating a NumPy array from a list of scores
scores = np.array([85, 92, 78, 99, 88])
print("Scores Array:", scores)

# Calculating the total sum of the scores
total = np.sum(scores)
print("Total Sum:", total) 

# Calculating the average score
average = np.mean(scores)
print("Average Score:", average) 
```

```python
#          ----------------------------------          Standard deviation           -------------------------------
# In Python, standard deviation is a measure of how spread out the values in a dataset are from the mean. It is calculated using the formula:
# std = sqrt(sum((x_i - mean)^2) / n)
# where x_i represents each value in the dataset, mean is the average of the dataset, and n is the number of values in the dataset.

# Calculating the standard deviation of the scores
std_dev = np.std(scores)
print("Standard Deviation:", std_dev) 
```
