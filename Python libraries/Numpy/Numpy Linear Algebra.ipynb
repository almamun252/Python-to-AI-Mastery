Linear Algebra

```python
import numpy as np
```

```python
#   ---------------------------- Dot Product (Matrix Multiplication) ----------------------------

inputs = np.array([[1.0, 2.0]]) # shape: (1, 2)
weights = np.array([[0.2, 0.8],
                    [0.5, 0.1]]) # shape: (2, 2)

# Dot product of inputs and weights
# output = np.dot(inputs, weights) or np.matmul(inputs, weights)
output = inputs @ weights 

print("Output of Matrix Multiplication:\n", output)
```

```python
#   ---------------------------- Transpose of a Matrix ----------------------------

matrix = np.array([[1, 2, 3],
                   [4, 5, 6]]) # shape: (2, 3)

transposed = matrix.T # or np.transpose(matrix)

print("Original Shape:", matrix.shape)
print("Transposed Shape:", transposed.shape) # shape: (3, 2)
print("Transposed Matrix:\n", transposed)
```

```python
#   ---------------------------- Inverse of a Matrix ----------------------------

A = np.array([[4, 7],
              [2, 6]])

# Calculate the inverse of A
A_inv = np.linalg.inv(A)
print("Inverse of A:\n", A_inv)
```

```python
# -------------------------------- Determinant of a Matrix ----------------------------

det_A = np.linalg.det(A)
print("Determinant of A:", round(det_A)) 
```

```python
#   ---------------------------- L2 Norm of a Vector ----------------------------
# Regularization (L1 or L2 Regularization) is used to prevent the model from overfitting during training in AI models. 
# It basically measures the length or distance of a vector.

vector = np.array([3, 4])

# L2 Norm (This is the square root of the sum of squares: sqrt(3^2 + 4^2) = 5)
norm_l2 = np.linalg.norm(vector)
print("L2 Norm:", norm_l2)
```
