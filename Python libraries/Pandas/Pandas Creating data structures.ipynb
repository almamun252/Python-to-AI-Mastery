Series & DataFrame

```python
import pandas as pd
import numpy as np
```

```python
# -------------------------------          Creating a Pandas Series from a List           -------------------------------

# A Python list of fruits
fruits_list = ["Apple", "Banana", "Mango", "Orange"]

# Creating a Pandas Series from the list
fruits_series = pd.Series(fruits_list)

print(fruits_series)
```

```python
# -------------------------------          Creating a Pandas Series from a Dictionary           -------------------------------

# A Python dictionary of fruits
fruits_dict = {
    "A": "Apple",
    "B": "Banana",
    "M": "Mango",
    "O": "Orange"
}

# Creating a Pandas Series from the dictionary
series_from_dict = pd.Series(fruits_dict)

print("Dictionary from which the Series is created:\n")
print(series_from_dict)
```

```python
# -------------------------------          Creating a Pandas Series from a NumPy Array           -------------------------------


# A NumPy array of fruits
fruits_array = np.array(["Apple", "Banana", "Mango", "Orange"])

# Creating a Pandas Series from the NumPy array
series_from_array = pd.Series(fruits_array)

print("NumPy Array from which the Series is created:\n")
print(series_from_array)

# You can also specify a custom index:
new_fruits_series = pd.Series(fruits_array, index=["A", "B", "M", "O"])
print("\nSeries with Custom Index:")
print(new_fruits_series)
```

DataFrame: 2D Data - Table/Rows and Columns

```python
# -------------------------------          Creating a Pandas DataFrame from a Dictionary           -------------------------------

# A dictionary, where keys are column names and values are the data
people_data = {
    "Name": ["Rahim", "Karim", "Mina", "Raju"],
    "Age": [25, 22, 24, 26],
    "City": ["Dhaka", "Sylhet", "Chattogram", "Rajshahi"]
}

# Creating a DataFrame from the dictionary
df = pd.DataFrame(people_data)

print(df)
```

```python
# -------------------------------          Creating a Pandas DataFrame from a List of Lists           -------------------------------

# A list of lists, where each inner list represents a row of data
people_list = [
    ["Rahim", 25, "Dhaka"],
    ["Karim", 22, "Sylhet"],
    ["Mina", 24, "Chattogram"],
    ["Raju", 26, "Rajshahi"]
]

# Column names for the DataFrame
column_names = ["Name", "Age", "City"]

# Creating a DataFrame from the list of lists
df_from_list = pd.DataFrame(people_list, columns=column_names)

print("DataFrame created from the list of lists:\n")
print(df_from_list)
```

```python
# -------------------------------          Creating a Pandas DataFrame from a NumPy Array           -------------------------------

# A 2D NumPy Array
people_array = np.array([
    ["Rahim", 25, "Dhaka"],
    ["Karim", 22, "Sylhet"],
    ["Mina", 24, "Chattogram"],
    ["Raju", 26, "Rajshahi"]
])

# Column names for the DataFrame
column_names = ["Name", "Age", "City"]

# Creating a DataFrame from the NumPy array
df_from_array = pd.DataFrame(people_array, columns=column_names)

print("DataFrame created from the NumPy array:\n")
print(df_from_array)
```

```python
# -------------------------------          Update DataFrame and Add new Column           -------------------------------

# DataFrame
df = pd.DataFrame({'Item': ['Apple', 'Banana'], 'Price': [100, 50]})
print("Original DataFrame:\n", df)

# new column 'Discount' added with values
df['Discount'] = [10, 5] 

# old column 'Price' updated with new values
df['Price'] = [120, 60] 

print("\nUpdated DataFrame:\n", df)
```
