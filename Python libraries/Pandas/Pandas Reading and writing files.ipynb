Reading and writing files

```python
import pandas as pd
import sqlite3
```

For CSV File

```python
#  ---------------------------------------- Create DataFrame And Write to CSV ----------------------------------------


# ১. Create a dictionary with the data
data = {
    'Date': ['2026-03-01', '2026-03-02', '2026-03-03', '2026-03-04'],
    'Store': ['Dhaka', 'Chittagong', 'Dhaka', 'Sylhet'],
    'Quantity': [50, 30, 45, 20],
    'Price_per_kg': [180, 185, 180, 190]
}

# ২. Dictionary to DataFrame
df = pd.DataFrame(data)

# Write the DataFrame to a CSV file
df.to_csv('apple_sells.csv', index=False) # index=False to avoid writing row numbers to the CSV file

print("DataFrame has been written to 'apple_sells.csv' successfully.")
print(df)
```

```python
#  ---------------------------------------- Read CSV, Update DataFrame, and Write Back to CSV ----------------------------------------
# read CSV file
df_read = pd.read_csv('apple_sells.csv')

# Add a new column for total revenue (Quantity * Price)
df_read['Total_Revenue'] = df_read['Quantity'] * df_read['Price_per_kg']

# Write the updated DataFrame back to a new CSV file
df_read.to_csv('updated_apple_sells.csv', index=False)

print("\nUpdated DataFrame:")
print(df_read)
```

For Excel File

```python
#  ---------------------------------------- Create DataFrame And Write to Excel ----------------------------------------

# ১. Create a dictionary with the data (laptop sales data)
data = {
    'Date': ['2026-03-01', '2026-03-02', '2026-03-03', '2026-03-04'],
    'Brand': ['HP', 'Dell', 'HP', 'Lenovo'],
    'Units_Sold': [10, 8, 15, 5],
    'Price_per_Unit': [50000, 55000, 50000, 45000]
}

# ২. Dictionary to DataFrame
df = pd.DataFrame(data)

# Write the DataFrame to an Excel file (with a specific sheet name)
df.to_excel('laptop_sales.xlsx', sheet_name='March_Sales', index=False) # index=False to avoid writing row numbers to the Excel file


print("DataFrame has been written to 'laptop_sales.xlsx' successfully.")
print(df)

```

```python

#  ---------------------------------------- Read Excel, Update DataFrame, and Write Back ----------------------------------------

# Read the specific sheet from the Excel file
df_read = pd.read_excel('laptop_sales.xlsx', sheet_name='March_Sales')

# Add a new column for total sales revenue (Units_Sold * Price_per_Unit)
df_read['Total_Sales'] = df_read['Units_Sold'] * df_read['Price_per_Unit']

# Write the updated DataFrame back to a NEW Excel file
df_read.to_excel('updated_laptop_sales.xlsx', sheet_name='Final_Report', index=False)

print("\nUpdated DataFrame (Total Sales column added):")
print(df_read)
```

```python
#  ---------------------------------------- Analyze Data (Groupby) ----------------------------------------

# Group data by 'Brand' and calculate the sum of 'Total_Sales'
# Brand wise total sales revenue calculated by grouping the data based on 'Brand' and summing up the 'Total_Sales' for each brand
brand_revenue = df_read.groupby('Brand')['Total_Sales'].sum().reset_index() # reset_index() is used to convert the resulting Series back into a DataFrame for better readability
print("\nBrand wise total sales revenue:")
print(brand_revenue)

# Sort the result to see which brand made the most money (Descending order)
brand_revenue_sorted = brand_revenue.sort_values(by='Total_Sales', ascending=False).reset_index(drop=True)

print("\nBrand wise total sales revenue sorted (Descending order):")
print(brand_revenue_sorted)
```

For JESON File

```python
#  ---------------------------------------- Create DataFrame And Write to JSON ----------------------------------------

# ১. Create a list of dictionaries (representing online orders data)
data = [
    {'Date': '2026-03-01', 'Category': 'Electronics', 'Quantity': 2, 'Unit_Price': 15000},
    {'Date': '2026-03-02', 'Category': 'Clothing', 'Quantity': 5, 'Unit_Price': 1200},
    {'Date': '2026-03-03', 'Category': 'Electronics', 'Quantity': 1, 'Unit_Price': 25000},
    {'Date': '2026-03-04', 'Category': 'Grocery', 'Quantity': 10, 'Unit_Price': 150}
]

# ২. List of Dictionaries to DataFrame
df = pd.DataFrame(data)

# Write the DataFrame to a JSON file
# orient='records' to write the JSON file as a list of dictionaries, and indent=4 for pretty printing
df.to_json('online_orders.json', orient='records', indent=4)

print("DataFrame has been written to 'online_orders.json' successfully.\n")
print(df)

```

```python

#  ---------------------------------------- Read JSON, Update DataFrame, and Write Back ----------------------------------------

# Read the JSON file
df_read = pd.read_json('online_orders.json')

# Add a new column for total amount (Quantity * Unit_Price)
df_read['Total_Amount'] = df_read['Quantity'] * df_read['Unit_Price']

# Write the updated DataFrame back to a NEW JSON file (with 'iso' date format)
df_read.to_json('updated_online_orders.json', orient='records', indent=4, date_format='iso')

print("\nUpdated DataFrame (Total Amount column added):")
print(df_read)

```

```python

#  ---------------------------------------- Analyze Data (Groupby) ----------------------------------------

# Group data by 'Category' and calculate the sum of 'Total_Amount'
# According to 'Category', the data is grouped and their 'Total_Amount' is summed up
category_sales = df_read.groupby('Category')['Total_Amount'].sum().reset_index()

# Sort the result to see which category made the most money (Descending order)
category_sales_sorted = category_sales.sort_values(by='Total_Amount', ascending=False).reset_index(drop=True)

print("\nTotal Sales by Category:")
print(category_sales_sorted)
```

For SQL Database

```python
#  ---------------------------------------- Create DataFrame And Write to SQL Database ----------------------------------------

# ১. Create a dictionary with the data (employee data)
data = {
    'Emp_ID': [101, 102, 103, 104, 105],
    'Name': ['Rahim', 'Karim', 'Mina', 'Raju', 'Sadia'],
    'Department': ['IT', 'HR', 'IT', 'Sales', 'HR'],
    'Salary': [60000, 45000, 75000, 50000, 48000]
}

# ২. Dictionary to DataFrame
df = pd.DataFrame(data)

# ৩. Create a connection to a local SQLite database
# If the file 'company_data.db' does not exist, Python will automatically create it when we connect to it.
conn = sqlite3.connect('company_data.db')

# ৪. Write the DataFrame to a SQL Table named 'employees'
# if_exists='replace' to replace the table if it already exists, and index=False to avoid writing row numbers to the SQL table
df.to_sql('employees', conn, index=False, if_exists='replace')

print("DataFrame has been successfully written to the 'employees' table in SQL database.\n")
print(df)
```

```python
#  ---------------------------------------- Read SQL, Update DataFrame, and Write Back ----------------------------------------

# ১. Read data using an SQL Query
query = "SELECT * FROM employees"
df_read = pd.read_sql(query, conn)

# ২. Add a new column for Eid Bonus (let's say, 10% of the original salary as Eid Bonus)
df_read['Eid_Bonus'] = df_read['Salary'] * 0.10

# ৩. Write the updated DataFrame back to a NEW SQL Table named 'employees_with_bonus'
df_read.to_sql('employees_with_bonus', conn, index=False, if_exists='replace')

print("\nUpdated DataFrame (Eid Bonus column added):")
print(df_read)

```

```python

#  ---------------------------------------- Analyze Data (Groupby) ----------------------------------------

# ১. Group data by 'Department' and calculate the total Salary per department
# 'Department' wise total salary calculated by grouping the data based on 'Department' and summing up the 'Salary' for each department
dept_salary = df_read.groupby('Department')['Salary'].sum().reset_index()

# ২. Sort the result (Descending order)
dept_salary_sorted = dept_salary.sort_values(by='Salary', ascending=False)

print("\nTotal Salary by Department:")
print(dept_salary_sorted)

# after finishing the work, it's good practice to close the database connection
conn.close()
```
